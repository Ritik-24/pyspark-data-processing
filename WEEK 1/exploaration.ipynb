```python
df['final_price'] = df['final_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
df.to_csv('cleaned_dataset.csv', index=False, encoding='utf-8-sig')
df[['title', 'initial_price', 'final_price']]
```

```python
#for calculating the final price after discount and total amount after discount
df['discount'] = pd.to_numeric(df['discount'], errors='coerce').fillna(0)
df['final_price'] = (df['initial_price'] * (1 - (df['discount'] / 100))).round(2)
#using the rating column to determine the quantity of each product, assuming higher ratings indicate more purchases
df['rating'] = pd.to_numeric(df['rating'], errors='coerce').fillna(0)
df['quantity'] = df['rating'].round().astype(int).apply(lambda x: 1 if x == 0 else x)

df['total_amount'] = (df['final_price'] * df['quantity']).round(2)

df[['initial_price', 'discount', 'final_price', 'quantity', 'total_amount']].head()
```

```python
category_summary = df.groupby('category')['total_amount'].sum()
category_summary
```

```python
# To Remove Corrupted Blocks from Seller Information Columns
import re

def clear_corrupted_blocks(text):
    if isinstance(text, str):
        cleaned = re.sub(r'[^\x00-\x7F]+', '', text)
        cleaned = re.sub(r'[\x00-\x1F\x7F]', '', cleaned)
        return cleaned.strip()
    return text

columns_to_clean = ['seller_name', 'seller_info', 'seller_information']

for col in columns_to_clean:
    if col in df.columns:
        df[col] = df[col].apply(clear_corrupted_blocks)

df.to_csv('cleaned_dataset.csv', index=False, encoding='utf-8-sig')
```
